```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sqlalchemy import create_engine
```

```python
countries = pd.read_csv('../06_Interactive_Visualization/data/country_codes.csv')
countries
```

```python
from dotenv import dotenv_values

config=dotenv_values(".env")
```

```python
username = config['USER']
password = config['PASS']
host = config['HOST']
port = config['PORT']
```

```python
url = f'postgresql+psycopg2://{username}:{password}@{host}:{port}/climate'
```

```python
engine = create_engine(url, echo=False)
```

```python
with engine.begin() as conn:
    conn.execute("DROP TABLE IF EXISTS countries ;")
    conn.execute("""
        CREATE TABLE countries (
            index INT,
            name VARCHAR,
            alpha2 VARCHAR(2) primary key,
            alpha3 VARCHAR,
            code INT,
            lat INT,
            lOn NUMERIC
        );
    """)
    countries.to_sql ('countries', conn, if_exists='append', index=True)
```
