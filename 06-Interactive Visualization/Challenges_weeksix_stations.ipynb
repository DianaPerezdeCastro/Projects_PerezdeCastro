```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sqlalchemy import create_engine
```

```python
stations=pd.read_csv('../06_Interactive_Visualization/data/ECA_blend_tg/stations.txt',skiprows = 17, encoding = 'unicode_escape')
```

```python
stations
```

```python
stations.columns = stations.columns.str.lower().str.strip()
```

```python
stations
```

```python
from dotenv import dotenv_values

config=dotenv_values(".env")
```

```python
username = config['USER']
password = config['PASS']
host = config['HOST']
port = config['PORT']
```

```python
url = f'postgresql+psycopg2://{username}:{password}@{host}:{port}/climate'
```

```python
engine = create_engine(url, echo=False)
```

```python
with engine.begin() as conn:
    conn.execute("DROP TABLE IF EXISTS stations ;")
    conn.execute("""
        CREATE TABLE stations(
            index INT,
            staid INT,
            staname VARCHAR, 
            cn VARCHAR(2) references countries,
            lat VARCHAR,
            lon VARCHAR,
            hght NUMERIC
        );
    """)
    stations.to_sql ('stations', conn, if_exists='append', index=True)
```
