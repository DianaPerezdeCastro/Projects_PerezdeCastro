```python


# initiating Nominatim, add you email address





# function
def reverse_geocoding(row):
    try:
        location = geolocator.reverse(f"{row['reclat']}, {row['reclong']}")
        address = location.raw['address']
        country = address.get('country', '')
        return country
    except:
        return None

# testing the function
# reverse_geocoding(met_new.loc[0, ['reclat', 'reclong']])


# running only first 10 rows
#met_new['country'] = met_new.loc[0:10, ['reclat', 'reclong']].progress_apply(reverse_geocoding, axis=1)

# this would require ~6 hours of computer work
met_new['country'] = met_new[['reclat', 'reclong']].progress_apply(reverse_geocoding, axis=1)
```
