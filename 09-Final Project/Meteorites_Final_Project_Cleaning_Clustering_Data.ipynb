### Creating a category, another clustering for a more generic view

```python
dictionary_category={'chondrite':['chondrite_carbonaceus','chondrite_ordinary','chondrite_enstatite','chondrite_other'],
                     'achondrite':['achondrite'],
                     'pallasite':['pallasite'],
                     'mesosiderite': ['mesosiderite'],
                     'iron': ['iron']}
                     
                                                    
for key, value in dictionary_category.items():

    for name in value:
        
        met.loc[met['sub_category'] == name, 'category'] = key






```

## Creating the first group of classification. Stony, stony-iron and iron

```python
dictionary_group={'stony':['chondrite','achondrite'],
                     'stony_iron':['pallasite','mesosiderite'],
                     'iron': ['iron']}
                     
                                                    
for key, value in dictionary_group.items():

    for name in value:
        
        met.loc[met['category'] == name, 'group'] = key
```
